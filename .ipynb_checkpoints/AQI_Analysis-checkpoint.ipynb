# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from folium.plugins import MarkerCluster

# Optional: For better visuals
sns.set(style="whitegrid")
```

# Loading CSV File

```python
# Load the CSV file
file_path = r"C:\Users\aasee\Downloads\AQI.csv"
df = pd.read_csv(file_path)

# Convert 'last_update' to datetime format
df['last_update'] = pd.to_datetime(df['last_update'], format='%d-%m-%Y %H:%M')
```

# Data Overview

```python
# Display basic info
df.info()
```

```python
pollutant_counts = df['pollutant_id'].value_counts()
pollutant_counts
```

```python
top_polluted_cities = df.groupby('city')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
top_polluted_cities
```

#  Most Common Pollutants

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='pollutant_id', order=pollutant_counts.index)
plt.title('Most Common Pollutants')
plt.xlabel('Count')
plt.ylabel('Pollutant')
plt.tight_layout()
plt.show()
```

# Top 10 Polluted Cities

```python
plt.figure(figsize=(12, 6))
top_polluted_cities.plot(kind='bar', color='orange')
plt.title('Top 10 Polluted Cities (Average Pollutant Level)')
plt.ylabel('Average Pollutant Level')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Top 10 States with Highest Pollution

```python
top_states = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_states.values, y=top_states.index, palette="Blues_d")
plt.title('Top 10 Polluted States (Average Pollutant Level)')
plt.xlabel('Average Pollutant Level')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

# City-wise Pollutant Distribution (Boxplot)

```python
# Pick top 6 cities with most data points for clearer boxplot
top_cities_by_count = df['city'].value_counts().head(6).index
filtered_df = df[df['city'].isin(top_cities_by_count)]

plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, x='city', y='pollutant_avg', palette="pastel")
plt.title('Pollutant Level Distribution by City')
plt.xlabel('City')
plt.ylabel('Pollutant Average')
plt.tight_layout()
plt.show()
```

# Heatmap of Average Pollution per City and Pollutant

```python
# Pivot to get a matrix of average pollutant levels
city_pollutant_matrix = df.pivot_table(values='pollutant_avg', index='city', columns='pollutant_id', aggfunc='mean')

# Reduce to top cities only for readability
top_cities_heat = df['city'].value_counts().head(10).index
city_pollutant_matrix = city_pollutant_matrix.loc[top_cities_heat]

plt.figure(figsize=(12, 6))
sns.heatmap(city_pollutant_matrix, annot=True, fmt=".1f", cmap="YlOrBr")
plt.title("Average Pollutant Levels in Top 10 Cities")
plt.xlabel("Pollutant")
plt.ylabel("City")
plt.tight_layout()
plt.show()
```

# Pie Chart – Pollutant Distribution by Frequency

```python
pollutant_counts = df['pollutant_id'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(pollutant_counts, labels=pollutant_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Pollutant Type Distribution (by Number of Readings)')
plt.axis('equal')  # Equal aspect ratio makes the pie round.
plt.tight_layout()
plt.show()
```

# Geo Map using Folium (State-wise Average Pollution)

```python
# Drop rows without coordinates
map_df = df.dropna(subset=['latitude', 'longitude'])

# Get state-wise average pollution with coordinates
map_group = map_df.groupby('state').agg({
    'pollutant_avg': 'mean',
    'latitude': 'first',
    'longitude': 'first'
}).reset_index()

# Create map centred in India
india_map = folium.Map(location=[22.9734, 78.6569], zoom_start=5)
marker_cluster = MarkerCluster().add_to(india_map)

# Add markers
for _, row in map_group.iterrows():
    popup_text = f"{row['state']}<br>Avg Pollution: {row['pollutant_avg']:.2f}"
    folium.CircleMarker(
        location=[row['latitude'], row['longitude']],
        radius=7,
        popup=popup_text,
        color='red',
        fill=True,
        fill_color='red'
    ).add_to(marker_cluster)

# Display the map
india_map
```

# Dashboard

```python
import dash
from dash import html, dcc
import pandas as pd
import plotly.express as px
import folium
from folium.plugins import MarkerCluster
import io
from dash.dependencies import Input, Output
import base64

# Load dataset
df = pd.read_csv(r"C:\Users\aasee\Downloads\AQI.csv")  # Replace with your file name
df['last_update'] = pd.to_datetime(df['last_update'], errors='coerce')

# Clean data
df = df.dropna(subset=['pollutant_avg'])

# Dash App
app = dash.Dash(__name__)
app.title = "Pollution Dashboard"

# Pie chart: pollutant distribution
pollutant_pie = px.pie(
    df,
    names='pollutant_id',
    title='Pollutant Type Distribution',
    hole=0.3
)

# Bar chart: top 10 polluted states
top_states = df.groupby('state')['pollutant_avg'].mean().sort_values(ascending=False).head(10)
state_bar = px.bar(
    x=top_states.values,
    y=top_states.index,
    orientation='h',
    title='Top 10 Polluted States',
    labels={'x': 'Average Pollution', 'y': 'State'},
    color=top_states.values,
    color_continuous_scale='Blues'
)

# Boxplot: pollutant spread across top cities
top_cities = df['city'].value_counts().head(6).index
box_data = df[df['city'].isin(top_cities)]
city_box = px.box(
    box_data,
    x='city',
    y='pollutant_avg',
    title='Pollutant Distribution by City',
    color='city'
)

# Heatmap
pivot_df = df.pivot_table(values='pollutant_avg', index='city', columns='pollutant_id', aggfunc='mean')
pivot_df = pivot_df.loc[df['city'].value_counts().head(10).index]
heatmap = px.imshow(
    pivot_df,
    text_auto=True,
    color_continuous_scale='YlOrBr',
    title='City vs Pollutant Heatmap'
)

# Folium Map (converted to iframe)
def create_map():
    map_df = df.dropna(subset=['latitude', 'longitude'])
    grouped = map_df.groupby('state').agg({
        'pollutant_avg': 'mean',
        'latitude': 'first',
        'longitude': 'first'
    }).reset_index()

    india_map = folium.Map(location=[22.9734, 78.6569], zoom_start=5)
    marker_cluster = MarkerCluster().add_to(india_map)

    for _, row in grouped.iterrows():
        folium.CircleMarker(
            location=[row['latitude'], row['longitude']],
            radius=7,
            popup=f"{row['state']}<br>Avg Pollution: {row['pollutant_avg']:.2f}",
            color='red',
            fill=True,
            fill_color='red'
        ).add_to(marker_cluster)

    map_buffer = io.BytesIO()
    india_map.save(map_buffer, close_file=False)
    return map_buffer.getvalue().decode()

map_html = create_map()

# Layout
app.layout = html.Div([
    html.H1("Air Pollution Data Dashboard", style={'textAlign': 'center'}),
    
    dcc.Graph(figure=pollutant_pie),
    dcc.Graph(figure=state_bar),
    dcc.Graph(figure=city_box),
    dcc.Graph(figure=heatmap),

    html.H3("Pollution Map of India", style={'textAlign': 'center'}),
    html.Iframe(srcDoc=map_html, width='100%', height='600')
])

# Run
if __name__ == '__main__':
    app.run(debug=True)
```

#  Least Polluted States

```python
import matplotlib.pyplot as plt
import pandas as pd

# Assuming `df` is already defined and contains 'state' and 'pollutant_avg' columns
least_states = df.groupby('state')['pollutant_avg'].mean().sort_values().head(10)

# Plotting with matplotlib
plt.figure(figsize=(10, 6))
bars = plt.barh(least_states.index, least_states.values, color='mediumseagreen')
plt.xlabel('Average Pollution')
plt.ylabel('State')
plt.title('Least Polluted States (Bottom 10)')
plt.gca().invert_yaxis()  # Optional: so the least polluted appears at the top

# Add value labels on the bars
for bar in bars:
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f}', va='center')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group by city and get bottom 10 based on average pollution
least_cities = df.groupby('city')['pollutant_avg'].mean().sort_values().head(10)

# Plotting with matplotlib
plt.figure(figsize=(10, 6))
bars = plt.barh(least_cities.index, least_cities.values, color='mediumpurple')
plt.xlabel('Average Pollution')
plt.ylabel('City')
plt.title('Least Polluted Cities (Bottom 10)')
plt.gca().invert_yaxis()  # To show least polluted at the top

# Add value labels on the bars
for bar in bars:
    plt.text(bar.get_width() + 0.1, bar.get_y() + bar.get_height()/2,
             f'{bar.get_width():.2f}', va='center')

plt.tight_layout()
plt.show()
```
